```python
import pykat
import pylab as pl
% matplotlib inline
```

```python
pykat = reload(pykat)
kat = pykat.finesse.kat()
kat.loadKatFile('check_jitter_mod.kat')

out=kat.run(printerr=1)
```

```python
pl.figure()
pl.subplot(211)
pl.plot(out.x,abs(out['adHG00USB']),label='HG00 USB')
#pl.plot(out.x,abs(out['adHG10USB']),label='HG10 USB')
pl.plot(out.x,abs(out['adHG01USB']),label='HG01 USB')
pl.plot(out.x,abs(out['adHG00LSB']),label='HG00 LSB')
#pl.plot(out.x,abs(out['adHG10LSB']),label='HG10 LSB')
pl.plot(out.x,abs(out['adHG01LSB']),label='HG01 LSB')
pl.legend()
pl.ylabel('Field amplitude [$\sqrt{W}$ ]')
pl.ylim((-0.01,0.045))
pl.tight_layout
pl.grid()
pl.subplot(212)
pl.plot(out.x,180/pl.pi*pl.angle(out['adHG00USB']),label='HG00 USB')
#pl.plot(out.x,180/pl.pi*pl.angle(out['adHG10USB']),label='HG10 USB')
pl.plot(out.x,180/pl.pi*pl.angle(out['adHG01USB']),label='HG01 USB')
pl.plot(out.x,180/pl.pi*pl.angle(out['adHG00LSB']),label='HG00 LSB')
#pl.plot(out.x,180/pl.pi*pl.angle(out['adHG10LSB']),label='HG10 LSB')
pl.plot(out.x,180/pl.pi*pl.angle(out['adHG01LSB']),label='HG01 LSB')
pl.legend()
pl.xlabel('EOBD modulation depth [rad]')
pl.ylabel('Field phase [deg]')
pl.tight_layout
pl.grid()

```
